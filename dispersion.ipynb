```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams["font.size"] = 25
```

```python
k = np.linspace(0, 20, 200)
VA = 1.0
di = 0.1
```

```python
def omegaplus(k):
    t1 = np.sqrt((k*di)**2 + 4)
    t2 = k*VA
    t3 = k*di
    return (t1+t3)*t2*0.5

def omegaminus(k):
    t1 = np.sqrt((k*di)**2 + 4)
    t2 = k*VA
    t3 = k*di
    return (t1-t3)*t2*0.5
```

```python
omegap = omegaplus(k)
omegam = omegaminus(k)
```

```python
plt.plot(k, omegap, "k", label=r'$\omega_+$', lw=2.5)
plt.plot(k, omegam, "r", label=r'$\omega_-$', lw=2.5)
plt.xlabel(r'$k$')
plt.ylabel(r'$\omega$')
plt.title(r'Dispersion relation for $V_A = 1.0$ and $d_i = 0.1$')
plt.legend()
plt.tight_layout()
plt.savefig("dispersion.jpg")
```
